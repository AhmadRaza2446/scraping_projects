```python
import pandas as pd
import time
from scrapy import Selector
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
```

```python
df = pd.read_csv('tripadvisor top40 cities.csv')
df
```

```python
list_of_dict = list()
for url in df['urls']:
    driver.get(url)
    res = Selector(text=driver.page_source)
    product_page_urls = res.css('a.bHGqj.Cj.b::attr(href)').getall()
    
    for index,product_page in enumerate(product_page_urls):
        if index == 20:
            break
        else:
            temp = dict()
            url = 'https://www.tripadvisor.de' + product_page
            driver.get(url)
            res1 = Selector(text=driver.page_source)
            name = res1.css('h1.fHibz::text').get()
            address = res1.css('span.dUpPX:nth-child(1) > span:nth-child(1) > a:nth-child(2)::text').get()
            phone = res1.css('span.fhGHT > a:nth-child(1)::text').get()
            website = res1.css('span.dUpPX:nth-child(3) > span:nth-child(1) > a:nth-child(2)::attr(href)').get()

            print('Name: ', name)
            print('Address: ', address)
            print('Phone: ', phone)
            print('Website: ', website)

            temp = dict()
            temp['name'] = name
            temp['address'] = address
            temp['phone'] = phone
            temp['website'] = website

            list_of_dict.append(temp)


df = pd.DataFrame(list_of_dict)
df.to_csv('data.csv',index=False)
     
```
