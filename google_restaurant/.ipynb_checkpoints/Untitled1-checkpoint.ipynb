```python
import pandas as pd
```

```python
df = pd.read_csv('final.csv')
df
```

```python
import numpy as np
df['total_reviews'] = df['total_reviews'].fillna(value=np.random.randn())
```

```python
df['total_reviews'] = df['total_reviews'].str.replace(' reviews','')
df['total_reviews'] = df['total_reviews'].str.replace(' review','')
```

```python
df['rating'] = df['rating'].astype(float)
df
```

```python
df.to_csv('final_v1.csv',index=False)
```
