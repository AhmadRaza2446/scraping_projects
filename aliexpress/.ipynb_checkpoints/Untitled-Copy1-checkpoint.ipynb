```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from scrapy import Selector
import pandas as pd
import time
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.maximize_window()
```

# Girls Clothing

```python
girls_list = list()
for page_num in range(60,61):
    print('Page Num: ', page_num)
    url = 'https://www.aliexpress.com/category/100003199/girls-clothing.html?trafficChannel=main&catName=girls-clothing&CatId=100003199&ltype=wholesale&SortType=default&page={}'.format(page_num)
    driver.get(url)
    time.sleep(3)
    for i in range(0,12):
        html = driver.find_element(By.CSS_SELECTOR,'a._3t7zg._2f4Ho')
        html.send_keys(Keys.PAGE_DOWN)
        time.sleep(2)

    res = Selector(text=driver.page_source)

    products_page_links = res.css('div.JIIxO a._3t7zg._2f4Ho::attr(href)').getall()

    
    product_number = 1
    for link in products_page_links:
        try:
            print('Product Number',product_number)
            product_number += 1
            temp = dict()
            link = 'https:'+link
    #         print(link)
            driver.get(link)
            time.sleep(1)
#             WebDriverWait(driver, timeout=10).until(
#             EC.visibility_of_element_located((By.CSS_SELECTOR, "span.sku-title-value"))
#             )
            res1 = Selector(text=driver.page_source)
            name = res1.css('h1.product-title-text::text').get()
    #         print('Name: ', name)
            price = res1.css('span.product-price-value::text').get()
            if not price:
                price = res1.css('span.uniform-banner-box-price::text').get()
    #         print('Price: ', price)
            images = res1.css('div.sku-property-image img::attr(src)').getall()
            if images:
                images = ','.join(images)
    #         print('Images: ', images)

            sizes = res1.css('div.sku-property-text span::text').getall()
            if sizes:
                sizes = ','.join(sizes)
            else:
                sizes = None
    #         print('Size: ', sizes)

            color = res1.css('span.sku-title-value::text').get()
    #         print('Color: ', color)

            temp['name'] = name
            temp['price'] = price
            temp['images'] = images
            temp['size'] = sizes
            temp['color'] = color
            temp['category'] = 'girls_clothing'

            print('Temp: ', temp)

            girls_list.append(temp)
        except:
            driver.refresh()
            time.sleep(1)
#             WebDriverWait(driver, timeout=10).until(
#             EC.visibility_of_element_located((By.CSS_SELECTOR, "span.sku-title-value"))
#             )
            res1 = Selector(text=driver.page_source)
            name = res1.css('h1.product-title-text::text').get()
    #         print('Name: ', name)
            price = res1.css('span.product-price-value::text').get()
            if not price:
                price = res1.css('span.uniform-banner-box-price::text').get()
    #         print('Price: ', price)
            images = res1.css('div.sku-property-image img::attr(src)').getall()
            if images:
                images = ','.join(images)
    #         print('Images: ', images)

            sizes = res1.css('div.sku-property-text span::text').getall()
            if sizes:
                sizes = ','.join(sizes)
            else:
                sizes = None
    #         print('Size: ', sizes)

            color = res1.css('span.sku-title-value::text').get()
    #         print('Color: ', color)

            temp['name'] = name
            temp['price'] = price
            temp['images'] = images
            temp['size'] = sizes
            temp['color'] = color
            temp['category'] = 'girls_clothing'

            print('Temp: ', temp)

            girls_list.append(temp)

girls_df = pd.DataFrame(girls_list)
girls_df
```

```python
girls_df = pd.DataFrame(girls_list)
# girls_df
# girls_df = girls_df[:180]
# girls_df
girls_df.to_csv('girls9.csv',index=False)
```
